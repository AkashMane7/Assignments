# PCA


```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
sc= StandardScaler()

x=sc.fit_transform(df)
x 
```

```python
pca = PCA()
pcs = pca.fit_transform(x)
pcs
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=3)
pcs = pca.fit_transform(x)
pcs
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
data = pd.DataFrame(pcs, columns=['pc1','pc2','pc3'])
data['Type']=df['Type']
```

```python
data
```

```python
plt.figure(figsize=(15,13))
sns.scatterplot(data=data)
```

# Hierarchical Clustering


```python
from scipy.cluster import hierarchy
from sklearn.preprocessing import normalize
from sklearn.cluster import AgglomerativeClustering
```

```python
df_norm=pd.DataFrame(normalize(df),columns=df.columns)
```

```python
df_norm.head(6)
```

```python
lk= hierarchy.linkage(df_norm, method='single')
dendo= hierarchy.dendrogram(lk)
```

```python
lk= hierarchy.linkage(df_norm, method='complete')
dendo= hierarchy.dendrogram(lk)
plt.axhline(y=0.2, c='red', linestyle='--' )
```

# Model building


```python
hc= AgglomerativeClustering(n_clusters=3)

y=hc.fit_predict(df_norm)
```

```python
y
```

```python
df['Cluster']= y
df.head(6)
```

```python
df.groupby('Cluster').agg('mean')
```

```python
plt.scatter(df['Alcohol'],df['Cluster'], c=y)
```

# K-means


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x= sc.fit_transform(df)
x 
```

```python
df_sc=pd.DataFrame(x ,columns=df.columns)
df_sc.head(5)
```

```python
km= KMeans()
y= km.fit_predict(x)
y 
```

```python
df1 = pd.DataFrame(x, columns=df.columns)
df1['cluster'] = y
df1.head(5)
```

```python
km.cluster_centers_
```

# Elbow Graph


```python
wcss=[]

for i in range (1,21):
    km= KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,21), wcss)
plt.xlabel('K-Value')
plt.ylabel('WCSS')
plt.title('Elbow Curve')
```

# Final Model


```python
km= KMeans(n_clusters=3)
y=km.fit_predict(x)
y 
```

```python
df['Cluster']=y
df.head(6)
```

```python
df.groupby('Cluster').agg('mean')
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['Cluster'], df['Alcohol'], c=km.labels_)
```
